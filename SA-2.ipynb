```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')
```

```python
df = pd.read_csv("/Users/vismaya/Downloads/Reviews.csv")
```

```python
print(df.head())
```

```python
stop_words = set(stopwords.words('english'))

def clean_text(text):
    tokens = word_tokenize(str(text).lower())
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words]
    return " ".join(tokens)
```

```python
df['clean_review'] = df['Text'].apply(clean_text)
```

```python
sid = SentimentIntensityAnalyzer()
df['sentiment_score'] = df['clean_review'].apply(lambda x: sid.polarity_scores(x)['compound'])
df['sentiment_label'] = df['sentiment_score'].apply(lambda x: 'positive' if x > 0.05 else ('negative' if x < -0.05 else 'neutral'))
```

```python
print(df['sentiment_label'].value_counts())
```

```python
sns.countplot(x='sentiment_label', data=df)
plt.title("Sentiment Distribution")
plt.show()
```

```python
positive_words = " ".join(df[df['sentiment_label']=='positive']['clean_review']).split()
positive_freq = pd.Series(positive_words).value_counts()[:20]
positive_freq.plot(kind='bar', title='Top Positive Words')
plt.show()
```
